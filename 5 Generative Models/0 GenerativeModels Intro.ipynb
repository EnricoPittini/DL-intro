# GenerativeModels Intro

Brief introduction about generative models.

### Discriminative vs Generative models

In ML, we can divide models into two classes.
- **Discriminative models** They are simply interested in drawing a separation line, for dividing the classes guessing. They are simply interested in guessing the label. Example: logistic regression. These models are used only for supervised learning.
- **Generative models** Their aim is to understand the distribution of the data. This is more powerfult than simply drwaing a separation line: with the distribution of data, you can do more. Indeed, these models can be used noth for supervised and unsupervised learning. Example: Naive Bayes, which is used for supervised learning.



**Probability distribution of data**

We have our features (for an image, the features are the pixels). Given some specific values for these features, how are these values likely? How likely is that this combination of values is "valid" and "sensible" with respect to the dataset?

Example. Specific image: value for each pixel. How likely is that this image represents a face?

This is the probability distributuon of data. $P(X)$, where $X$ is our instance, i.e. collection of features.

### Generative models

So, a generative model is a model which tries to learn the actual distribution of data (of the training set).

Why is that interesting?
- For acquiring information on the data
- For generating new data, which are similar to the data you already have.

In general, generative models are interesting in multi-modal tasks, namely tasks in which there are more possible outputs.

### How the probability distribution is given

Two possible ways.
- Implicit way. A **generator** is built, which returns a sample from the distribution $P(X)$. It samples from $P(X)$.
- Explicit way. The explicit probability distribution is found.

We will see only the implicit way.

## LATENT VARIABLES MODELS

All the generative models that we will see (i.e. VAE and GAN) are laten variables models.

The idea is simple: we generate an instance $\hat{X}$ starting froma latent variable $z$, which is a vector of values (vector of latent variables). $z$  can be called *latent encoding*, and it can be seen as a starting random seed/noise, useful for our generation.

### Latent space

We suppose to know the probability distribution $P(z)$. This is the latent space. Space of the latent variable $z$.

This space/probability distribution is typically a multivariate normal gaussian ($0$ mean, unitarian variance). $N(0,1)$.

### Marginalization of $P(X)$

We can express $P(X)$ through marginalization over $z$.

$P(X) = \int P(X|z)*P(z)$. 

Basically, we express $P(X)$ by means of the conditional probability of $X$ given $z$ and by means of the probability of $z$. $P(X|z)$ is the **posterior**, $P(z)$ the prior.

### Sampling from $P(X)$

The posterior $P(X|z)$ allows us to generate a sample $\hat{X}$ given $z$. It is a **generator**.

We generate a new instance $\hat{X}$ starting from $z$. We conditionally generate $\hat{X}$ from $z$

### Why do we need to generate $\hat{X}$ starting from a latent variable $z$?

In other words, why do we need the seed $z$?

Because the generator $P(X|z)$ is implemeneted as a NN, which is a deterministic function. So, if we start from nothing, we always generate the same instance $\hat{X}$. For this reason, I conditionally start from a seed. From a random variable.
