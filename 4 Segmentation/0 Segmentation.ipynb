# Segmenetation

Let's briefly see what segmentation is in general.

### Semantic segmenetation

Given an image, we try to classify each individual pizel. We have a fixed set of possible categories: for each pixel, we want to associate a category. It's a kind of classification problem, but at pixel level.



WIth semantic segmentation, we are trying to associate a meaning. We try to understand the image. We are not trying to classify the main object in the image (like in classification) but we are trying to identify and classify the different portions of the image.

### Semantic segmenetation: image-to-image processing

The output is still an image: so, semantic segmentation is a kind of image-to-image processing. We are translating an image into another image, representing the semantic of the input image: each output pixel carries out a category.

This is good for NNs: an image-to-image task can be addressed by means of a single NN, in an end-to-end fashion.

So, the output is a single image, in which each pixel contains a category, a class. So, single channel. 

### Output image: insights

Actually, the output is an image which contains, for each pixel, the probability distribution of that pixel over the classes (i.e. categorical distribution). For each pixel, we have the categorical distribution over the classes. Therefore, if we have $k$ classes, for each pixel we have $k$ values/probabilities. This means that the output is an image with $k$ channels, one for each class.The $i$-th channels contains the probabilities of belonging to the corresponding $i$-th class. 

Given the output image with $k$ channels, we can easily get the output image with only $1$ channels, in which each pixel is simply associated to a single class. For doing this, we simply take the most probable class of each pixel.
